```python
import matplotlib.pyplot as plt
import numpy as np
import simpful as sf
from functools import reduce
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
import openpyxl
```

```python
# Inicialização
FS = sf.FuzzySystem(show_banner=False)

# Entrada: Numero de Pessoas
NP1 = sf.FuzzySet(function=sf.Triangular_MF(0, 0, 5), term="Less")
NP2 = sf.FuzzySet(function=sf.Triangular_MF(2, 5, 8) , term="Normal")
NP3 = sf.FuzzySet(function=sf.Triangular_MF(5, 10, 10), term="More")
numeroPessoas = sf.LinguisticVariable([NP1,NP2,NP3], universe_of_discourse=[0,10])
FS.add_linguistic_variable("Numero_de_Pessoas", LV=numeroPessoas)

# Entrada: Tamanho do Quarto
RS1 = sf.FuzzySet(function=sf.Triangular_MF(0, 0, 100), term="Small")
RS2 = sf.FuzzySet(function=sf.Triangular_MF(40, 100, 160) , term="Medium")
RS3 = sf.FuzzySet(function=sf.Triangular_MF(100, 200, 200), term="Large")
tamanhoQuarto = sf.LinguisticVariable([RS1,RS2,RS3], universe_of_discourse=[0,200])
FS.add_linguistic_variable("Tamanho_do_Quarto", LV=tamanhoQuarto)

# Entrada: Temperatura Atmosferica
TA1 = sf.FuzzySet(function=sf.Trapezoidal_MF(-100, -100, 15, 30), term="Cold")
TA2 = sf.FuzzySet(function=sf.Triangular_MF(22.5, 30, 37.5) , term="Warm")
TA3 = sf.FuzzySet(function=sf.Trapezoidal_MF(30, 45, 100, 100), term="Hot")
temperaturaAtmosferica = sf.LinguisticVariable([TA1,TA2,TA3], universe_of_discourse=[-100,100])
FS.add_linguistic_variable("Temperatura_Atmosferica", LV=temperaturaAtmosferica)

# Saida: Temperatura do Quarto
TQ1 = sf.FuzzySet(function=sf.Trapezoidal_MF(4.5,4.5,16,23), term='Low')
TQ2 = sf.FuzzySet(function=sf.Triangular_MF(18.5,23,27.5), term='Medium')
TQ3 = sf.FuzzySet(function=sf.Trapezoidal_MF(23,30,37,37), term='High')
temperaturaQuarto = sf.LinguisticVariable([TQ1,TQ2,TQ3], universe_of_discourse=[4.5,37])
FS.add_linguistic_variable("Temperatura_Quarto", LV=temperaturaQuarto)

# Saida: Capacidade de Refrigeração
CP1 = sf.FuzzySet(function=sf.Trapezoidal_MF(2400,2400,3500,4600), term='VeryCold')
CP2 = sf.FuzzySet(function=sf.Triangular_MF(4050,4850,5650), term='Cold')
CP3 = sf.FuzzySet(function=sf.Trapezoidal_MF(5375,6500,7625,7625), term='warm')
capacidadeRefrigeracao = sf.LinguisticVariable([CP1,CP2,CP3], universe_of_discourse=[2400,7625])
FS.add_linguistic_variable("Capacidade_Refrigeracao", LV=capacidadeRefrigeracao)

# Saida: Potencia consumida
PC1 = sf.FuzzySet(function=sf.Trapezoidal_MF(700,700,1075,1450), term='Low')
PC2 = sf.FuzzySet(function=sf.Triangular_MF(1250,1550,1830), term='Medium')
PC3 = sf.FuzzySet(function=sf.Trapezoidal_MF(1725,2000,2275,2275), term='High')
potenciaConsumida = sf.LinguisticVariable([PC1,PC2,PC3], universe_of_discourse=[700,2275])
FS.add_linguistic_variable("Potencia_Consumida", LV=potenciaConsumida)

# Saida: Inclinação
I1 = sf.FuzzySet(function=sf.Triangular_MF(0,0,12.5), term='Low')
I2 = sf.FuzzySet(function=sf.Triangular_MF(3.75,15,27.5), term='Centre')
I3 = sf.FuzzySet(function=sf.Triangular_MF(18.15,30,30), term='High')
inclinacao = sf.LinguisticVariable([I1,I2,I3], universe_of_discourse=[0,30])
FS.add_linguistic_variable("Inclinação", LV=inclinacao)

# Saida: Velocidade do Vento
VV1 = sf.FuzzySet(function=sf.Triangular_MF(600,600,840), term='Low')
VV2 = sf.FuzzySet(function=sf.Triangular_MF(660,900,1140), term='Medium')
VV3 = sf.FuzzySet(function=sf.Triangular_MF(960,1200,1200), term='Fast')
velocidadeVento = sf.LinguisticVariable([VV1,VV2,VV3], universe_of_discourse=[600,1200])
FS.add_linguistic_variable("velocidade_Vento", LV=velocidadeVento)
```

#### Criação de regras
___

As regras vão precisar ser escritas na seguinte maneira:

"**IF** (Entrada_1 **IS** ruim) **AND** (Entrada_2 **IS** bom) **THEN** (Saida **IS** medio)"

```python
df = pd.read_excel("regras.xlsx")
```

```python
df.columns
```

```python
df['COOLING CAPACITY'].unique()
```

```python
rule = []
for i, row in df.iterrows():
    rule.append(f"IF (Numero_de_Pessoas IS {row['NUMBER OF PERSON']}) AND (Tamanho_do_Quarto IS {row['ROOM SIZE']}) AND (Temperatura_Atmosferica IS {row['ATMOSSPHERIC TEMPERATURE']}) THEN (Capacidade_Refrigeracao IS {row['COOLING CAPACITY']})") # (Temperatura_Quarto IS {row['ROOM TEMPERATURE']}) (Capacidade_Refrigeracao IS {row['COOLING CAPACITY']}) (Potencia_Consumida IS {row['POWER CONSUMPTION']}) (Inclinação IS {row['TILTS POSITION']}) (velocidade_Vento IS {row['FAN SPEED']})")
FS.add_rules(rule)
```

```python
X_room_size = np.arange(0, 200, 8)
X_len = len(X_room_size)
Y_number_person = np.arange(0, 10, 0.4)
Y_len = len(Y_number_person)
Z_temo = np.arange(-100, 100, 20)
Z_len = len(Z_temo)

Z_decisao = np.zeros((X_len, Y_len))

# Loop para calcular os valores de z
for i in range(X_len):
    for j in range(Y_len):
        #for k in range(Z_len):
        FS.set_variable('Tamanho_do_Quarto', X_room_size[i])
        FS.set_variable('Numero_de_Pessoas', Y_number_person[j])
        #FS.set_variable('Temperatura_Atmosferica', Z_temo[k])
        Z_decisao[i, j] = FS.Mamdani_inference(subdivisions=1000)['Capacidade_Refrigeracao']
        print(FS.Mamdani_inference(subdivisions=1000))
```

```python
X, Y = np.meshgrid(X_room_size, Y_number_person)
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z_decisao.T, cmap='magma')

ax.set_xlabel('Tamanho_Quarto')
ax.set_ylabel('Numero_Pessoas')
ax.set_zlabel('Temperatura_Quarto')

plt.title('Superfície Fuzzy')
plt.tight_layout()
ax.view_init(azim=-120, elev=20)

plt.show()
```
